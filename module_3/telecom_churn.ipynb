```python
# 1.4. Определение типов признаков

numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

print("Числовых признаков:", len(numeric_features))
print("Категориальных признаков:", len(categorical_features))
```

Разделяем признаки на числовые и категориальные:

* Числовые (int/float) будут масштабироваться и заполняться медианой.  
* Категориальные (object) будут кодироваться с помощью one‑hot encoding и заполняться наиболее частым значением.

**Результаты и выводы:**

* Обнаружено 78 числовых и 20 категориальных признаков (вместе с `churn` исходно было 78+21, после отделения `churn` осталось 78+20).  
* Наличие значимого числа категориальных признаков оправдывает использование автоматического one‑hot кодирования через `ColumnTransformer`.

```python
# 1.5. Конвейер предобработки: импутация, масштабирование, кодирование

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown='ignore'))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

Строим единый конвейер предобработки:

* Для числовых признаков: медианная импутация + стандартизация.  
* Для категориальных: заполнение модой + one‑hot encoding с игнорированием новых категорий.

**Результаты и выводы:**

* Все шаги предобработки инкапсулированы в `preprocess`, что снижает риск ошибок и утечки данных между train и test.  
* При кросс‑валидации и на валидационном наборе будет использоваться одна и та же схема обработки данных, что делает эксперименты воспроизводимыми.

```python
# 2.5. Черновой анализ важности признаков с помощью случайного леса

tmp_rf_pipe = Pipeline(steps=[
    ("preprocess", preprocess),
    ("clf", RandomForestClassifier(
        n_estimators=200,
        random_state=42,
        n_jobs=-1,
        class_weight='balanced'
    ))
])

tmp_rf_pipe.fit(X, y)

cat_encoder = tmp_rf_pipe.named_steps['preprocess'].named_transformers_['cat']['onehot']
cat_feature_names = cat_encoder.get_feature_names_out(categorical_features)
num_feature_names = np.array(numeric_features)
all_feature_names = np.concatenate([num_feature_names, cat_feature_names])

feature_importances = tmp_rf_pipe.named_steps['clf'].feature_importances_
fi_tmp = pd.DataFrame({"feature": all_feature_names, "importance": feature_importances})
fi_tmp.sort_values(by="importance", ascending=False).head(20)
```

Обучаем простой случайный лес и смотрим важность признаков:

```python
# 3. Построение моделей классификации

# 3.1. Разбиение на train/test

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

```python
# 3.2. Вспомогательная функция оценки моделей

def evaluate_model(pipe, X_train, y_train, X_test, y_test, model_name="model"):
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    
    # Кросс-валидация по ROC-AUC
    y_train_proba_cv = cross_val_predict(
        pipe, X_train, y_train,
        cv=cv, method='predict_proba'
    )[:, 1]
    y_train_pred_cv = (y_train_proba_cv >= 0.5).astype(int)
    
    metrics_cv = {
        "accuracy_cv": accuracy_score(y_train, y_train_pred_cv),
        "precision_cv": precision_score(y_train, y_train_pred_cv),
        "recall_cv": recall_score(y_train, y_train_pred_cv),
        "f1_cv": f1_score(y_train, y_train_pred_cv),
        "roc_auc_cv": roc_auc_score(y_train, y_train_proba_cv),
    }
    
    # Обучение на train и оценка на test
    pipe.fit(X_train, y_train)
    y_test_proba = pipe.predict_proba(X_test)[:, 1]
    y_test_pred = (y_test_proba >= 0.5).astype(int)
    
    metrics_test = {
        "accuracy_test": accuracy_score(y_test, y_test_pred),
        "precision_test": precision_score(y_test, y_test_pred),
        "recall_test": recall_score(y_test, y_test_pred),
        "f1_test": f1_score(y_test, y_test_pred),
        "roc_auc_test": roc_auc_score(y_test, y_test_proba),
    }
    
    print(f"\n===== {model_name} =====")
    print("CV metrics:")
    for k, v in metrics_cv.items():
        print(f"{k}: {v:.4f}")
    print("Test metrics:")
    for k, v in metrics_test.items():
        print(f"{k}: {v:.4f}")
    
    ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred)
    plt.title(model_name + " confusion matrix (test)")
    plt.show()
    
    print("\nClassification report (test):")
    print(classification_report(y_test, y_test_pred))
    
    return pipe, metrics_cv, metrics_test
```

Функция `evaluate_model` унифицирует процесс оценки:

* Проводит 5‑fold стратифицированную кросс‑валидацию и вычисляет Accuracy, Precision, Recall, F1 и ROC‑AUC на train.  
* Обучает модель на train и оценивает её на test по тем же метрикам.  
* Строит матрицу ошибок и выводит `classification_report`.

**Результаты и выводы:**

* После вызова `evaluate_model` для разных моделей мы получаем сопоставимые наборы метрик.  
* Это позволяет объективно сравнить алгоритмы и не забыть про важные показатели (например, Recall и ROC‑AUC для класса оттока).

# 4. Тонкая настройка модели

Выбираем случайный лес как модель‑кандидат и настраиваем его гиперпараметры:

* Добавляем в конвейер `SMOTE` для балансировки классов на уровне обучающей выборки.  
* Настраиваем число деревьев, глубину, минимальное количество объектов в узлах, стратегию выбора признаков и веса классов.

* Средний ROC‑AUC на кросс‑валидации для лучшей модели выше, чем у базового леса, что оправдывает использование гиперпараметрической оптимизации и SMOTE.

```python
# 4. Тонкая настройка модели

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

imb_rf_pipe = ImbPipeline(steps=[
    ("preprocess", preprocess),
    ("smote", SMOTE(random_state=42)),
    ("clf", RandomForestClassifier(random_state=42, n_jobs=-1))
])

param_distributions = {
    "clf__n_estimators": randint(200, 600),
    "clf__max_depth": randint(3, 15),
    "clf__min_samples_split": randint(2, 10),
    "clf__min_samples_leaf": randint(1, 5),
    "clf__max_features": ["sqrt", "log2", 0.5, 0.7],
    "clf__class_weight": [None, "balanced"]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

rand_search = RandomizedSearchCV(
    estimator=imb_rf_pipe,
    param_distributions=param_distributions,
    n_iter=30,
    scoring='roc_auc',
    n_jobs=-1,
    cv=cv,
    verbose=2,
    random_state=42
)

rand_search.fit(X_train, y_train)

print("Best ROC-AUC (CV):", rand_search.best_score_)
print("Best params:", rand_search.best_params_)

best_model = rand_search.best_estimator_
```
