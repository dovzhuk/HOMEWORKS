# 14. Какая команда выиграла наибольшое количество матчей в сезоне? (Будьте внимательны с типом матча в сезоне - season).

```python
# учитываем только регулярный сезон
regular = df[df["season"] == "regular season"]

# победы дома
wins_home = regular.groupby("home_team")["home_team_win"].sum()

# победы в гостях (когда home_team_win == 0)
wins_away = regular.groupby("away_team")["home_team_win"].apply(lambda s: (1 - s).sum())

# суммарные победы
total_wins = wins_home.add(wins_away, fill_value=0)

# команда‑лидер и число побед
leader = total_wins.idxmax()
max_wins = int(total_wins.max())

answer = f"Ответ: наибольшее количество матчей выиграла команда {leader} с количеством в {max_wins} матча."
answer
```

# 17. Какая команда проиграла наибольшее количество матчей в сезоне?

```python
# Рассматриваем только матчи регулярного сезона (`season == "regular season"`), чтобы исключить иные типы игр.
# Группируем данные по `home_team` и суммируем по `home_team_loss` - находим поражения дома.
# Группируем данные по `away_team` и суммируем по `home_team_win` - находим поражения в гостях.
# Суммируем обе полученные серии.
# Из итоговой серии общих поражений по командам выбирается команда с максимальным числом поражений.


# поражения дома
losses_home = regular.groupby("home_team")["home_team_loss"].sum()

# поражения в гостях: когда выигрывает домашняя команда (home_team_win == 1)
losses_away = regular.groupby("away_team")["home_team_win"].sum()

# суммарные поражения
total_losses = losses_home.add(losses_away, fill_value=0)

losses_leader = total_losses.idxmax()
losses_max = int(total_losses.max())

answer = f"Ответ: Наибольшее количество матчей проиграла команда {losses_leader} с результатом {losses_max} матча."
answer
```

# 18. Зависит ли выигрыш от количества посетителей матча? (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.cov.html)

```python
# Смотрим ковариацию и корреляцию между числовым признаком `attendance` и бинарным признаком `home_team_win` (0/1).
# Приводим `home_team_win` к типу `int`
# Считаем ковариацию и корреляцию.

# берём только нужные столбцы и убеждаемся, что win — числовой 0/1
sub = df[["attendance", "home_team_win"]].copy()
sub["home_team_win"] = sub["home_team_win"].astype(int)

# ковариация
cov_matrix = sub.cov()
cov_att_win = cov_matrix.loc["attendance", "home_team_win"]

# корреляция
corr_matrix = sub.corr()
corr_att_win = corr_matrix.loc["attendance", "home_team_win"]

answer = f"Так как ковариация между количеством зрителей и признаком победы домашней команды составляет {cov_att_win:.2f}, а коэффициент корреляции Пирсона равен {corr_att_win:.2f} (близок к нулю), линейной зависимости выигрыша от посещаемости по данным практически нет."
answer
```

# 20. Правда ли что наибольшее количество ранов происходит в холодную погоду? (Холодной погодой считается погода ниже 0 градусов)

```python
# 0°C = 32°F
cold_threshold_f = 32

# признак холодной погоды
df["is_cold"] = df["temperature"] < cold_threshold_f

# число игр в холод и не в холод
games_per_group = df["is_cold"].value_counts()

# суммарное и среднее число ранов по группам
runs_stats = df.groupby("is_cold")["total_runs"].agg(["count", "sum", "mean"])

games_per_group, runs_stats
```

Игр с температурой ниже 32°F почти нет, а среднее и суммарное число ранов в таких матчах не превосходит показатели игр в более тёплую погоду, поэтому утверждение, что наибольшее количество ранов происходит именно в «холодную» погоду (ниже 0°C), не подтверждается.
