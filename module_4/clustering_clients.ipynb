### Проверить наличие пропущенных значений и обработать их

```python
# Простой вариант обработки:
# 1) Числовые признаки — заполняем медианой
# 2) Категориальные признаки — заполняем модой (наиболее частым значением)

# Разделим столбцы на числовые и объектные
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns

# Заполнение числовых признаков медианой
for col in num_cols:
    if df[col].isna().sum() > 0:
        df[col] = df[col].fillna(df[col].median())

# Заполнение категориальных признаков модой
for col in cat_cols:
    if df[col].isna().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

# Проверим, что пропусков не осталось
print(df.isna().sum())
```

**Вывод:**  
1. Изначально в датасете были пропуски, например, в признаке дохода `AnnualIncome` и, возможно, в некоторых других столбцах.  
2. Для простоты мы заполнили пропуски в числовых признаках медианой (устойчиво к выбросам), а в категориальных признаках — самым частым значением.  
3. После заполнения проверка `isna().sum()` показывает, что пропущенные значения отсутствуют, и данные готовы к дальнейшей обработке.  

### Произвести анализ признаков на аномальные значения, то есть выбросов, и обработать их

```python
# Список только числовых признаков
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
num_cols
```

```python
# Посмотрим описательную статистику по числовым признакам
df[num_cols].describe().T
```

```python
# Функция для поиска выбросов по правилу IQR (межквартильный размах)
def find_outliers_iqr(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return series[(series < lower) | (series > upper)]

# Посмотрим количество выбросов для ключевых числовых признаков
cols_to_check = ["AnnualIncome", "SpentOnWines", "SpentOnFruits",
                 "SpentOnMeat", "SpentOnFish", "SpentOnSweets", "SpentOnGold"]

for col in cols_to_check:
    if col in df.columns:
        outliers = find_outliers_iqr(df[col])
        print(col, "- выбросов:", outliers.shape[0])
```

```python
# Простой вариант обработки выбросов — "обрезка" (winsorization) по границам IQR

def clip_outliers_iqr(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return series.clip(lower=lower, upper=upper)

# Применим обрезку для основных денежных признаков
for col in cols_to_check:
    if col in df.columns:
        df[col] = clip_outliers_iqr(df[col])
```

```python
# Проверим статистику после обработки выбросов
df[cols_to_check].describe().T
```

**Вывод:**  
1. С помощью описательной статистики и правила IQR мы обнаружили, что в денежных признаках (доход, траты по категориям) присутствуют аномально большие значения.  
2. Для упрощённой обработки выбросов мы применили обрезку (clip) по нижней и верхней границе IQR, чтобы слишком экстремальные значения не «тянули» на себя масштаб при кластеризации.  
3. После обработки распределения признаков стали более «компактными», что делает модель кластеризации более устойчивой к единичным аномальным клиентам.

### Поработайте с категориальными признаками, некоторые категории в них избыточны

```python
# Посмотрим уникальные значения категориальных признаков
cat_cols = df.select_dtypes(include=["object"]).columns
for col in cat_cols:
    print(f"\n{col}:")
    print(df[col].value_counts())
```

```python
# Укрупнение уровней образования
# Пример: объединим уровни в более крупные группы
education_map = {
    "Basic": "Basic",
    "2n Cycle": "Undergraduate",
    "Graduation": "Undergraduate",
    "Master": "Postgraduate",
    "PhD": "Postgraduate",
    "Absurd": "Other"  # странное/шумовое значение
}

if "EducationLevel" in df.columns:
    df["EducationLevel_grouped"] = df["EducationLevel"].map(education_map)

# Укрупнение семейного положения
marital_map = {
    "Single": "Single",
    "Alone": "Single",
    "Married": "Couple",
    "Together": "Couple",
    "Divorced": "Separated",
    "Widow": "Separated",
    "Absurd": "Other"  # если встретится
}

if "MaritalStatus" in df.columns:
    df["MaritalStatus_grouped"] = df["MaritalStatus"].map(marital_map)

# Проверим новые группировки
if "EducationLevel_grouped" in df.columns:
    print("\nEducationLevel_grouped:")
    print(df["EducationLevel_grouped"].value_counts())

if "MaritalStatus_grouped" in df.columns:
    print("\nMaritalStatus_grouped:")
    print(df["MaritalStatus_grouped"].value_counts())
```

**Вывод:**  
1. В исходных категориальных признаках `EducationLevel` и `MaritalStatus` присутствуют детальные и местами «шумовые» категории (например, `Absurd`, разделение похожих статусов), что усложняет модель и не всегда полезно для кластеризации [file:4].  
2. Мы укрупнили уровни образования в несколько более понятных групп (`Basic`, `Undergraduate`, `Postgraduate`, `Other`) и объединили семейные статусы в группы (`Single`, `Couple`, `Separated`, `Other`), тем самым уменьшили количество уникальных значений и упростили интерпретацию сегментов [web:22].  
3. Такие укрупнённые категории помогут алгоритмам кластеризации лучше выделять осмысленные группы клиентов и уменьшат риск переобучения на редких значениях.

## 2.1. Выполнить кластеризацию клиентов, используя разные методы

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

```python
# Будем кластеризовать матрицу признаков X (или X_pca, если используете PCA)
# Здесь используем X (без понижения размерности, чтобы было нагляднее)
X_clust = X

# 1) K-Means (зададим, например, 4 кластера как стартовый вариант)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_clust)

# 2) Иерархическая кластеризация (Agglomerative Clustering)
agglo = AgglomerativeClustering(n_clusters=4, linkage="ward")
agglo_labels = agglo.fit_predict(X_clust)

# 3) DBSCAN (параметры нужно будет подбирать; здесь простой стартовый вариант)
dbscan = DBSCAN(eps=2.0, min_samples=10)
dbscan_labels = dbscan.fit_predict(X_clust)

# Сохраним результаты кластеризации во временные столбцы
df_clusters = df.copy()
df_clusters["Cluster_KMeans_4"] = kmeans_labels
df_clusters["Cluster_Agglo_4"] = agglo_labels
df_clusters["Cluster_DBSCAN"] = dbscan_labels

df_clusters[[
    "CustomerID",
    "Cluster_KMeans_4",
    "Cluster_Agglo_4",
    "Cluster_DBSCAN"
]].head()
```

3. Метки кластеров сохранены в датафрейм `df_clusters`, позже мы выберем наилучший метод и число кластеров, оценим качество кластеризации и запишем финальный кластер в основной датасет.

## 2.2. Определить оптимальное количество кластеров

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Будем использовать матрицу признаков X_clust (как и раньше)
X_clust = X

inertias = []
sil_scores = []
k_values = range(2, 11)  # проверим k от 2 до 10

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_clust)
    
    inertias.append(kmeans.inertia_)  # сумма квадратов расстояний до центров
    sil_scores.append(silhouette_score(X_clust, labels))

# График метода локтя (inertia)
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(k_values, inertias, marker="o")
plt.title("Метод локтя (Inertia)")
plt.xlabel("Количество кластеров k")
plt.ylabel("Inertia")

# График силуэта
plt.subplot(1, 2, 2)
plt.plot(k_values, sil_scores, marker="o")
plt.title("Коэффициент силуэта")
plt.xlabel("Количество кластеров k")
plt.ylabel("Silhouette score")

plt.tight_layout()
plt.show()
```

**Вывод:**  
1. Метод «локтя» по графику Inertia помогает выбрать такое k, после которого снижение ошибки (Inertia) замедляется — эта точка считается хорошим кандидатом для количества кластеров.  
2. Коэффициент силуэта показывает, насколько хорошо объекты отделены друг от друга: чем выше значение, тем лучше разделены кластеры; сравнивая значения для разных k, можно выбрать k с максимальным или близким к максимальному силуэтом.  
3. На практике часто выбирают k, которое одновременно находится в «локте» графика Inertia и даёт достаточно высокий silhouette score, далее это значение используют для финальной кластеризации.
4. По графику метода локтя видно, что после k = 4 снижение Inertia замедляется, «локоть» находится примерно в области 4–5 кластеров.  
5. Коэффициент силуэта максимален при k = 2, но далее значения для k = 3–5 остаются достаточно высокими и близкими друг к другу.  
6. С учётом баланса между интерпретируемостью (больше 2 сегментов) и качеством разделения, разумным выбором для дальнейшей работы является 4–5 кластеров; в дальнейшем можно попробовать оба варианта и сравнить их при профилировании.  

## 2.3. Оцените качество полученной кластеризации

```python
# Возьмём в качестве базового варианта k = 4 (по результатам локтя/силуэта)
X_clust = X

kmeans_4 = KMeans(n_clusters=4, random_state=42, n_init=10)
labels_4 = kmeans_4.fit_predict(X_clust)

sil_4 = silhouette_score(X_clust, labels_4)
print("Silhouette score для K-Means (k=4):", sil_4)

# Для сравнения посчитаем silhouette для k = 3 и k = 5
for k in [3, 5]:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    lab = km.fit_predict(X_clust)
    sil = silhouette_score(X_clust, lab)
    print(f"Silhouette score для K-Means (k={k}): {sil}")
```

**Вывод:**  
1. Коэффициент силуэта для выбранного числа кластеров (k = 4) показывает, насколько хорошо объекты внутри кластеров похожи друг на друга и отделены от других кластеров: значения ближе к 1 означают хорошее разделение, около 0 — слабое, отрицательные — плохое.  
2. Сравнение значений силуэта для k = 3, 4 и 5 показывает, что качество кластеризации при k = 4 сопоставимо с соседними значениями и остаётся на достаточно хорошем уровне, что подтверждает выбор этого k как разумного компромисса между качеством и числом сегментов.  
3. Дальше будем использовать кластеризацию K-Means с k = 4 в качестве основной и сохранять полученные метки в датасет для последующего профилирования кластеров.  

## 2.4. Проведите кластеризацию наилучшим образом и сохраните результат кластеризации в новый столбец датасета

```python
# Используем матрицу признаков X и выбранное ранее количество кластеров k = 4
X_clust = X

best_k = 4
kmeans_final = KMeans(n_clusters=best_k, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(X_clust)

# Сохраняем метки кластеров в основной датасет
df["Cluster"] = cluster_labels

# Посмотрим на распределение клиентов по кластерам
df["Cluster"].value_counts().sort_index()
```

- Персонализированные e‑mail/SMS‑кампании с индивидуальными скидками на любимые категории и ранний доступ к новым продуктам; бонусные баллы за онлайн‑покупки и участие в акциях.  

**Кластер 3 — «Семьи со школьниками/подростками со средним доходом»:**  
- Семейные и сезонные акции: наборы для школьников, товары для завтраков и перекусов, комплекты «на выходные/каникулы», чтобы покрывать регулярные семейные потребности.  
- Комбинированные предложения: скидка при покупке набора категорий (например, мясо + овощи + сладости), кэшбэк или дополнительные бонусы за покупки в нескольких каналах (онлайн + офлайн), что стимулирует рост частоты и суммы покупок.

### Скидка для многодетных семей для клиентов с высоким числом детей и активными покупками через каталоги

**Кластер 1 — «Молодые семьи с детьми»:**  
- Специальные скидки для многодетных: повышенный процент скидки на базовые продукты и детские товары (молочные продукты, каши, подгузники), особенно при покупках через каталог или онлайн, где можно оформить крупный заказ сразу.  
- Пакетные предложения «для семьи»: большие упаковки, семейные наборы по сниженной цене, дополнительные бонусы за заказы по каталогу (например, бесплатная доставка при сумме заказа выше порога), что стимулирует активные каталог‑покупки.  

**Кластер 3 — «Семьи со школьниками/подростками»:**  
- Каталожные акции для семей: скидки на большие наборы продуктов «на неделю», наборы для школьников (перекусы, напитки), бонусные баллы за регулярные заказы по каталогу — удобно планировать закупки на семью.  
- Программы лояльности «семейный бонус»: дополнительные баллы/скидка за количество детей (по данным `NumChildren` и `NumTeenagers`) и за частоту каталог‑покупок, чтобы усилить привязку именно этого канала.  

**Кластеры 0 и 2 (менее многодетные, но ценные по тратам):**  
- Для **кластера 0** — семейные наборы со смешанными категориями (повседневные продукты + товары «для удовольствия»), небольшие бонусы за оформление заказов через каталог (доставка, промокоды).  
- Для **кластера 2** — премиальные каталог‑предложения (эксклюзивные наборы, подарочные боксы), персональные скидки на высокомаржинальные категории при заказе по каталогу, так как этот сегмент готов платить за удобство и сервис.

### Специальные предложения для любителей вина, если кластер выделяется высоким расходом на вино

**Кластер 2 — премиальные клиенты с высокими тратами (в т.ч. на вино):**  
- Персональные скидки на премиальные вина и дегустационные наборы: акции «-X% на любимые вина месяца», наборы из нескольких бутылок с бонусом, подарочные боксы — это соответствует их высокому чеку и ориентации на качество.  
- Эксклюзивные предложения: приглашения на дегустации, закрытые распродажи вин, ранний доступ к новинкам винной линейки, дополнительные бонусные баллы за покупку вина онлайн или по каталогу.  

**Кластеры 0 и 3 — умеренные любители вина:**  
- Пакетные предложения «ужин + вино»: скидка при покупке вина вместе с мясом/сыром/деликатесами, небольшие акции на вино в выходные, чтобы стимулировать дополнительные покупки без сильного давления на бюджет.  
- Персональные купоны на конкретные винные бренды, которые клиент уже покупал, с небольшим процентом скидки или кэшбэком, чтобы укреплять лояльность и поощрять повторные покупки.  

**Кластер 1 — клиенты с низкими тратами (в т.ч. на вино):**  
- Небольшие акции «попробуй выгодно»: недорогие позиции вина со скидкой, маленькие бутылки/мини‑наборы по сниженной цене, чтобы аккуратно стимулировать пробные покупки без сильной нагрузки на бюджет.  
- Комбинированные акции: небольшая скидка на вино при покупке определённой суммы базовых продуктов — это увязывает удовольствие (вино) с основной корзиной и не вызывает ощущения «дорогой роскоши».  

## 4.2. Описать, как эти рекомендации помогут улучшить маркетинговую стратегию компании

1. **Повышение точности таргетинга.**  
   Персональные предложения по кластерам (для премиальных, семей с детьми, экономных клиентов) позволяют направлять нужный товар и формат скидки тем, кто с наибольшей вероятностью откликнется, вместо «одинаковых» массовых акций.  

2. **Рост выручки и среднего чека.**  
   Для кластеров с высоким доходом и тратами (например, премиального сегмента) специальные офферы на премиальные продукты, вино и эксклюзивные сервисы стимулируют дополнительное потребление и увеличивают средний чек без сильного давления на маржу.  

3. **Оптимизация маркетинговых затрат.**  
   Для «холодных» сегментов с низким откликом (молодые семьи с низкими тратами) можно снижать частоту дорогих персональных кампаний и использовать более дешёвые, массовые форматы, перераспределяя бюджет в пользу более отзывчивых кластеров.  

4. **Укрепление лояльности и удержание клиентов.**  
   Семейные программы, предложения для многодетных, наборы «для недели» и комфортные каталог‑/онлайн‑сервисы делают взаимодействие с компанией более удобным и ценным для разных типов семей, снижая риск ухода к конкурентам.  

5. **Развитие омниканальной стратегии.**  
   Понимание того, какие кластеры активно используют онлайн и каталог, позволяет точечно развивать эти каналы (дополнительные бонусы, удобные акции), а не инвестировать одинаково во все направления; это усиливает эффект от цифровых инициатив.  

В итоге сегментированный подход на основе кластеров превращает маркетинг из «массовой рассылки» в управляемую систему: компания лучше понимает, **кто** её клиенты, **что** им важно и **как** эффективнее с ними коммуницировать, что ведёт к росту выручки и эффективности бюджета.
